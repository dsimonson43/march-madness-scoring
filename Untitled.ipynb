```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Define Round Name & Number of games per round
rounds = ["Round of 64", "Round of 32", "Sweet 16", "Elite 8", "Final Four", "Championship"]
games_per_round = [32, 16, 8, 4, 2, 1]
```

```python
#Scoring systems
scoring_systems = {
    "Standard": [1, 2, 4, 8, 16, 32],
    "Linear": [1, 2, 3, 4, 5, 6],
    "Fibonacci": [1, 2, 4, 7, 11, 16],
    "Custom": [1.0, 1.6, 2.2, 2.8, 3.4, 4.0]
}
```

```python
#create a dataframe for total possible PPR (Points per round) and total 
data = {"Round": rounds, "Games": games_per_round}
df = pd.DataFrame(data)
```

```python
#adding total points per scoring system
for name, values in scoring_systems.items():
    df[name] = [g * v for g, v in zip(games_per_round, values)]
```

```python
#calculating the percentage of total for each system
for name in scoring_systems:
    total_points = df[name].sum()
    df[f"{name} %"] = df[name] / total_points * 100
```

```python
#plotting the percentage of contribution per round
plt.figure(figsize=(12,6))
for name in scoring_systems:
    plt.plot(df["Round"], df[f"{name} %"], marker='o', label=name)

plt.title("Percentage of Total Points per Round by Scoring System")
plt.ylabel("Percentage of Total Points (%)")
plt.xlabel("Tournament Round")
plt.legend(title="Scoring System")
plt.grid(True)
plt.tight_layout()

df, plt.show()
```

```python
#New Dataframe for points per game//value per pick
ppg_data = {
    "Round": rounds,
    "Games": games_per_round
}
```

```python
#Adding each scoring system's per-game value directly\
for name, values in scoring_systems.items():
    ppg_data[name] = values

ppg_df = pd.DataFrame(ppg_data)
```

```python
#Plotting points per game per round
plt.figure(figsize=(12, 6))
for name in scoring_systems:
    plt.plot(ppg_df["Round"], ppg_df[name], marker='o', label=name)
plt.title("Points Per Game Per Round by Scoring System")
plt.ylabel("Points per correct pick")
plt.xlabel("Tournament Round")
plt.legend(title="scoring system")
plt.grid(True)
plt.tight_layout()

ppg_df, plt.show()
```
